```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime, timedelta
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# Fetch live data for a given cryptocurrency ticker
def fetch_crypto_data(ticker="EOS-USD", start=None, interval="1h"):
    if start is None:
        start = datetime.now() - timedelta(days=30)  # Default to last 30 days
    data = yf.Ticker(ticker).history(start=start, interval=interval)
    data = data[["Open", "High", "Low", "Close", "Volume"]]
    data["Close"] = pd.to_numeric(data["Close"], errors="coerce")
    data.dropna(inplace=True)
    return data

# Compute RSI (Relative Strength Index)
def compute_rsi(data, window=14):
    diff = data.diff(1)
    up_chg = diff.clip(lower=0)
    down_chg = -diff.clip(upper=0)
    up_chg_avg = up_chg.ewm(span=window, min_periods=window).mean()
    down_chg_avg = down_chg.ewm(span=window, min_periods=window).mean()
    rs = up_chg_avg / down_chg_avg
    return 100 - (100 / (1 + rs))

# Compute Bollinger Bands
def compute_bollinger_bands(data, window=20):
    rolling_mean = data.rolling(window=window).mean()
    rolling_std = data.rolling(window=window).std()
    upper_band = rolling_mean + 2 * rolling_std
    lower_band = rolling_mean - 2 * rolling_std
    return upper_band, lower_band

# Add technical indicators to the dataset
def add_technical_indicators(data):
    data["SMA"] = data["Close"].rolling(window=20).mean()
    data["EMA"] = data["Close"].ewm(span=20, adjust=False).mean()
    data["RSI"] = compute_rsi(data["Close"])
    data["Bollinger_Upper"], data["Bollinger_Lower"] = compute_bollinger_bands(data["Close"])
    data.dropna(inplace=True)
    return data

# Preprocess the data: Scale the features
def preprocess_data(data):
    data = add_technical_indicators(data)
    features = ["Open", "High", "Low", "Close", "Volume", "SMA", "EMA", "RSI", "Bollinger_Upper", "Bollinger_Lower"]
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data[features])
    return scaled_data, scaler

# Prepare the last data for prediction
def prepare_last_data(scaled_data, window_size=100):
    return scaled_data[-window_size:].reshape(1, window_size, scaled_data.shape[1])

# Predict future prices using the trained model
def predict_prices(model, last_data, scaler, hours=72):
    predicted_prices = []
    current_input = last_data
    num_features = last_data.shape[2]  # Get the number of features

    for _ in range(hours):
        next_price = model.predict(current_input)
        predicted_prices.append(next_price[0][0])

        # Prepare the next input by appending the predicted value
        next_price_reshaped = np.repeat(next_price.reshape(1, 1, 1), num_features, axis=2)
        current_input = np.append(current_input[:, 1:, :], next_price_reshaped, axis=1)

    # Convert scaled predictions back to actual prices
    future_prices_actual = []
    for price in predicted_prices:
        dummy_array = np.zeros((1, num_features))
        dummy_array[0, 3] = price  # Assume 'Close' is the 4th column (index 3)
        actual_price = scaler.inverse_transform(dummy_array)[0, 3]
        future_prices_actual.append(actual_price)

    return np.array(future_prices_actual)

# Plot the actual and predicted prices
def plot_prices(data, predicted_prices, hours=72):
    historical_data = data["Close"].iloc[-100:]
    historical_dates = data.index[-100:]
    predicted_dates = pd.date_range(start=historical_dates[-1] + timedelta(hours=1), periods=hours, freq="H")

    plt.figure(figsize=(12, 6))
    plt.plot(historical_dates, historical_data, label="Actual Prices", color="blue", marker="o")
    plt.plot(predicted_dates, predicted_prices, label="Predicted Prices", color="green", linestyle="--")
    plt.title("Cryptocurrency Prices: Actual and Predicted")
    plt.xlabel("Date & Time")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Main function to predict and plot cryptocurrency prices
def predict_and_plot_crypto_prices(ticker="EOS-USD", model_path="enhanced_crypto_lstm_model.h5", hours=72):
    # Fetch data
    data = fetch_crypto_data(ticker)

    # Preprocess data
    scaled_data, scaler = preprocess_data(data)

    # Load trained model
    model = load_model(model_path)

    # Prepare last data for prediction
    last_data = prepare_last_data(scaled_data)

    # Predict future prices
    future_prices = predict_prices(model, last_data, scaler, hours)

    # Plot results
    plot_prices(data, future_prices, hours)

    # Return predicted prices
    return pd.DataFrame({"Timestamp": pd.date_range(start=data.index[-1] + timedelta(hours=1), periods=hours, freq="H"),
                         "Predicted Price": future_prices})

# Example usage
if __name__ == "__main__":
    predicted_df = predict_and_plot_crypto_prices()
    print(predicted_df.head())

```
